from datetime import datetime

# Task class to store individual task details

class Task:
    def __init__(self, id, description):
        self.id = id
        self.description = description
        self.date = datetime.today().strftime("%Y-%m-%d")
        self.done = False

    def mark_done(self):
        self.done = True

    def __str__(self):
        return f"ID: {self.id} | {self.date} | Done: {self.done} | Description: {self.description}"


# TaskManager class to manage the task list

class TaskManager:
    def __init__(self):
        self.tasks = []
        self.task_id_counter = 1

    def add_task(self):
        description = input("Enter task description: ")
        task = Task(self.task_id_counter, description)
        self.tasks.append(task)
        print("✅ Task added!")
        self.task_id_counter += 1

    def list_tasks(self, mode="all"):
        filtered = []
        today = datetime.today().strftime("%Y-%m-%d")

        if mode == "done":
            filtered = [task for task in self.tasks if task.done]
        elif mode == "undone":
            filtered = [task for task in self.tasks if not task.done]
        elif mode == "pending":
            filtered = [task for task in self.tasks if not task.done and task.date < today]
        elif mode == "asc":
            filtered = sorted(self.tasks, key=lambda t: t.date)
        elif mode == "desc":
            filtered = sorted(self.tasks, key=lambda t: t.date, reverse=True)
        else:
            filtered = self.tasks

        if not filtered:
            print("⚠️ No tasks to show.")
        else:
            print("\n--- Task List ---")
            for task in filtered:
                print(task)
            print("------------------")

    def remove_task(self):
        try:
            task_id = int(input("Enter Task ID to remove: "))
            for task in self.tasks:
                if task.id == task_id:
                    self.tasks.remove(task)
                    print("🗑️ Task removed.")
                    return
            print("❌ Task not found.")
        except ValueError:
            print("⚠️ Invalid input. Please enter a valid number.")

    def mark_task_done(self):
        try:
            task_id = int(input("Enter Task ID to mark as done: "))
            for task in self.tasks:
                if task.id == task_id:
                    task.mark_done()
                    print("✅ Task marked as done.")
                    return
            print("❌ Task not found.")
        except ValueError:
            print("⚠️ Invalid input. Please enter a valid number.")

    def show_menu(self):
        print("\n====== Task Manager ======")
        print("1. Add Task")
        print("2. List All Tasks")
        print("3. List Done Tasks")
        print("4. List Undone Tasks")
        print("5. List Pending Tasks")
        print("6. List Tasks by Date (ASC)")
        print("7. List Tasks by Date (DESC)")
        print("8. Mark Task as Done")
        print("9. Remove Task")
        print("10. Exit")

    def run(self):
        while True:
            self.show_menu()
            try:
                choice = int(input("Enter your choice (1–10): "))
                if choice == 1:
                    self.add_task()
                elif choice == 2:
                    self.list_tasks()
                elif choice == 3:
                    self.list_tasks("done")
                elif choice == 4:
                    self.list_tasks("undone")
                elif choice == 5:
                    self.list_tasks("pending")
                elif choice == 6:
                    self.list_tasks("asc")
                elif choice == 7:
                    self.list_tasks("desc")
                elif choice == 8:
                    self.mark_task_done()
                elif choice == 9:
                    self.remove_task()
                elif choice == 10:
                    print("👋 Exiting Task Manager.")
                    break
                else:
                    print("⚠️ Invalid choice. Please choose from 1 to 10.")
            except ValueError:
                print("⚠️ Invalid input. Please enter a number.")
